```python
#  Histogram untuk kolom 'ikut serta dalam permusuhan'
plt.figure(figsize=(10, 6))
sns.histplot(df['took_part_in_the_hostilities'], bins=20, kde=False)
plt.title('Persentase Permusuhan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
df['date_of_event'] = pd.to_datetime(df['date_of_event'])
df['date_of_death'] = pd.to_datetime(df['date_of_death'])
```

```python
# Menghitung jumlah kejadian per hari
events_per_day = df.groupby('date_of_event').size()

plt.figure(figsize=(12, 6))
events_per_day.plot()
plt.title('Peristiwa Seiring Waktu')
plt.xlabel('Tanggal kejadian')
plt.ylabel('Jumlah Kejadian')
plt.show()
```

```python
events_by_killed_by = df.groupby(['date_of_event', 'killed_by']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
for column in events_by_killed_by.columns:
    plt.plot(events_by_killed_by.index, events_by_killed_by[column], label=column)

plt.title('Jumlah Peristiwa Seiring Waktu menurut "killed_by"')
plt.xlabel('Tanggal Kejadian')
plt.ylabel('Jumlah Kejadian')
plt.legend(loc='upper right', fontsize='small')
plt.xticks(rotation=45)
plt.show()
```

**Dari Grafik diatas kejadian pembunuhan banyak dilakukan oleh pasukan keamanan Israel**

```python
#Mengelompokkan kolom 'date_of_event' dan 'ammunition' serta menghitung jumlah kejadian untuk setiap kombinasi

events_by_ammunition = df.groupby(['date_of_event', 'ammunition']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
for column in events_by_ammunition.columns:
    plt.plot(events_by_ammunition.index, events_by_ammunition[column], label=column)

plt.title('Jumlah Kejadian Sepanjang Waktu berdasarkan "Amunisi"')
plt.xlabel('Tanggal Kejadian')
plt.ylabel('Jumlah Kejadian')
plt.legend(loc='upper right', fontsize='small')
plt.xticks(rotation=45)
plt.show()
```

Amunisi terbanyak yang dipakai pada kejadian adalah **roket qassam** dengan rentan kejadian pada 2008-2009
dan untuk kejadian akhir akhir ini kebanyakan amunisi yang dipakai adalah **tabung gas air mata dan juga misil**

```python
#Top 5 jumlah lokasi kejadian
top5_event_locations = df['event_location'].value_counts().head(5)

plt.figure(figsize=(12, 6))
sns.barplot(x=top5_event_locations.index, y=top5_event_locations.values)
plt.title('Top 5 jumlah lokasi kejadian')
plt.xlabel('Lokasi Kejadian')
plt.ylabel('Jumlah')
plt.xticks(rotation=90)
plt.show()
```
